```python
cpt = 'CPT W'
#1217
#import stuff
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.patches import PathPatch
import matplotlib as mpl
import pandas as pd

#load data
df = pd.read_csv('Y:/Data/CPTs/2012/ARECPT1217_2012-05-27_2346.csv',header=8,delimiter=';')

#load ScientificColourMaps colourmap (http://www.fabiocrameri.ch/colourmaps.php)
cm_data = np.loadtxt('Y:\Resources\ScientificColourMaps4\\lajolla\\lajolla.txt')
lajolla = LinearSegmentedColormap.from_list('CBname', cm_data)
lajolla_r = lajolla.reversed()


#Define x and y values
x = df.MPa.values
y = df.m.values

#resample x and y
new_y = np.linspace(y.min(), y.max(), len(y))
new_x = np.interp(new_y, y, x)

# Set up the figure.
fig, ax = plt.subplots(figsize=(4, 10))

# Make the background 'image'.
im = ax.imshow(new_x.reshape(-1,1),
               aspect='auto',
               origin='lower',
               extent=[0,60,new_y.min(),new_y.max()],
               cmap=lajolla_r
              )

# Draw the path.
paths = ax.fill_betweenx(y, x, x.min(),
                         facecolor='none',
                         lw=2,
                         edgecolor='black',
                        )

# Make the 'fill' mask and clip the background image with it.
patch = PathPatch(paths._paths[0], visible=False)
ax.add_artist(patch)
im.set_clip_path(patch)

# Finish up.
mpl.rc('font', family='Arial')
plt.xlabel('Cone resistance qc (MPa)',fontsize=26)
plt.xticks(np.arange(0,70,step=10),fontsize=22)
plt.ylabel('Depth (m)',fontsize=26)
plt.yticks(np.arange(0,y.max(),step=2),fontsize=22)
plt.minorticks_on()
plt.xlim(0,60)
plt.ylim(40,0)
ax.legend((cpt,''),loc=4,fontsize=26,handlelength=0)
plt.savefig(cpt)
plt.savefig(cpt, format='svg')
plt.show()
```
