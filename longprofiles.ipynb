```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rc('font',family='Arial')
```

```python
df = pd.read_csv('Y:/py/longprofiles.csv')
df.head()
```

```python
plt.plot(df.river1_main,df.r1m_s,color='#74a9cf',linewidth=0.5)
plt.plot(df.river1_trib0,df.r1t0_s,color='#0570b0',linewidth=0.5)
plt.plot(df.river1_trib1,df.r1t1_s,color='#023858',linewidth=0.5)
plt.xlim(0,20000)
plt.ylim(-64,-30)
plt.xlabel('profile distance (m)',fontsize=22)
plt.ylabel('elevation (m OD)',fontsize=22)
plt.xticks(np.arange(0,21000,step=1000),fontsize=18)
plt.yticks(np.arange(-64,-28,step=2),fontsize=18)
plt.legend(('main','trib1','trib2'),fontsize=22,loc=4)
plt.minorticks_on()
plt.title('Channel 1',fontsize=22,loc='left',pad=-20)
fig = plt.gcf()
fig.set_size_inches(20,5)
plt.tight_layout()
plt.savefig('river1.png')
plt.savefig('river1.svg')
```

```python
plt.plot(df.river2_main,df.r2m_s,color='#1d91c0',linewidth=0.5)
plt.plot(df.river2_trib0,df.r2t0_s,color='#225ea8',linewidth=0.5)
plt.plot(df.river2_trib1,df.r2t1_s,color='#253494',linewidth=0.5)
plt.plot(df.river2_trib2,df.r2t2_s,color='#081d58',linewidth=0.5)
plt.xlim(0,32000)
plt.ylim(-54,-30)
plt.xlabel('profile distance (m)',fontsize=22)
plt.ylabel('elevation (m OD)',fontsize=22)
plt.xticks(np.arange(0,33000,step=2000),fontsize=18)
plt.yticks(np.arange(-54,-28,step=2),fontsize=18)
plt.legend(('main','trib1','trib2','trib3'),fontsize=22,loc=4)
plt.minorticks_on()
plt.title('Channel 2',fontsize=22,loc='left',pad=-20)
fig = plt.gcf()
fig.set_size_inches(20,5)
plt.tight_layout()
plt.savefig('river2.png')
plt.savefig('river2.svg')
```
